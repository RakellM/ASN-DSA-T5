```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv("decathlon.csv", index_col=0)
df.head(30)
```

```python
# 1. Analise a matriz de correlações dos dados.

# Cálculo da matriz de correlações
corr = df.iloc[:,0:9].corr()
corr
```

```python
# Mapa de calor das correlações

sns.heatmap(corr, 
        xticklabels=corr.columns,
        yticklabels=corr.columns,  annot=True, linewidth=.4)
```

```python
# Análise de Componentes Principais
# 2. Decomponha a matriz de variâncias e covariâncias em componentes principais.


X = np.array(df.iloc[:,0:10])
S = np.cov(np.transpose(X))

S[0:2,0:9]
```

```python
# variâncias

np.diagonal(S)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=8)
pca
```

```python
pca.fit(X)
```

```python
np.round(pca.components_[0:2,:],3)
```

```python
# 3. Quantas componentes explicam ao menos 90% da variabilidade dos dados?

# Variância das componentes principais

pca.explained_variance_ratio_

np.round(pca.explained_variance_ratio_,3)
```

```python
pca.explained_variance_ratio_.cumsum()
```

```python
fig = plt.figure(figsize=(8,5))

plt.plot(pca.explained_variance_ratio_,  'ro-', linewidth=2)
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Eigenvalue')

plt.show()
```

```python
pca.transform(X)
```

```python
# Vamos considerar 2 componentes principais

pca = PCA(n_components=2)
pca.fit(X)

pca.components_
```

```python
np.round(pca.components_[1],3)
```

```python
PCA1 = pca.transform(X)[:,0]
PCA2 = pca.transform(X)[:,1]
```

```python
df['PCA1'] = PCA1
df['PCA2'] = PCA2
```

```python
df.head()
```

```python
df.sort_values(by='PCA1',ascending=True)
```

```python
df.sort_values(by='PCA2',ascending=False)
```

```python
from sklearn import preprocessing
```

```python
X_std = preprocessing.scale(X)
X_std = pd.DataFrame(X_std, columns =list(df.iloc[:,0:10]))
X_std.head()
```

```python

pca_std = PCA(n_components=5)
pca_std.fit(X_std)

np.round(pca_std.components_,3)
```

```python
np.round(pca_std.components_[0],3)
```

```python
np.round(pca_std.explained_variance_ratio_,2)
```

```python
fig = plt.figure(figsize=(8,5))

plt.plot(pca_std.explained_variance_ratio_,  'ro-', linewidth=2)
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Eigenvalue')

plt.show()
```

```python
pca_std.transform(X_std)
```

```python
PCA1 = pca_std.transform(X_std)[:,0]
PCA2 = pca_std.transform(X_std)[:,1]
```

```python
df['PCA_std1'] = PCA1
df['PCA2_std'] = PCA2
```

```python
df.sort_values(by='PCA_std1',ascending=True)
```

```python
scalePC1 = 1.0/(PCA1.max() - PCA1.min())
scalePC2 = 1.0/(PCA1.max() - PCA1.min())
features = list(df.iloc[:,0:10])

ldngs = pca_std.components_
```

```python
fig, ax = plt.subplots(figsize=(14, 9))
 
for i, feature in enumerate(features):
    ax.arrow(0, 0, ldngs[0, i], 
             ldngs[1, i])
    ax.text(ldngs[0, i] * 1.15, 
            ldngs[1, i] * 1.15, 
            feature, fontsize=18)
 
ax.scatter(PCA1 * scalePC1,PCA2 * scalePC2)
 
ax.set_xlabel('PCA1', fontsize=20)
ax.set_ylabel('PCA2', fontsize=20)
ax.set_title('Figure 1', fontsize=20)
```
